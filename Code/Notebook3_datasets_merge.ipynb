#### Extract Economic Feature Data from Coin Metrics Github

```python
#Read the csv file
df_AAVE=pd.read_csv('https://raw.githubusercontent.com/coinmetrics/data/master/csv/aave.csv')
df_AAVE['time']=pd.to_datetime(df_AAVE['time'])
df_AAVE.head()
```

```python
# define the time range for data analysis
start_existing = datetime.datetime(2020,10,10)
end_existing = datetime.datetime(2021,10,9)

start_new = datetime.datetime(2021,7,10)
end_new = datetime.datetime(2022,7,9)
```

```python
# select data during the defined time range
df_AAVE_existing = df_AAVE[(df_AAVE.time>=start_existing) & (df_AAVE.time<=end_existing)]
df_AAVE_existing.head()

# select data during the defined time range
df_AAVE_new = df_AAVE[(df_AAVE.time>=start_new) & (df_AAVE.time<=end_new)].reset_index(drop=True)
df_AAVE_new.head()
```

```python
# Load Network Feature data
Network_AAVE_new = pd.read_csv('Output_Data/AAVE_Network_Features_Updated.csv')
# move time to the first column
Network_AAVE_new.insert(0, 'time', Network_AAVE_new.pop('time'))
# drop columns
Network_AAVE_new = Network_AAVE_new.drop(Network_AAVE_new.columns[[1,2]], axis = 1)
# convert time to date
Network_AAVE_new['time']=pd.to_datetime(Network_AAVE_new['time'])
# sort data by time
Network_AAVE_new = Network_AAVE_new.sort_values(by='time', ascending=True).reset_index(drop=True)
# select time interval
Network_AAVE_new  = Network_AAVE_new[(Network_AAVE_new.time>=start_new) & (Network_AAVE_new.time<=end_new)]
# create relative degree column
Network_AAVE_new['relative_degree'] = Network_AAVE_new['num_edges']/(Network_AAVE_new['num_nodes']*(Network_AAVE_new['num_nodes']-1)/2)
Network_AAVE_new.head()
```

```python
# merge existing datasets
# df_AAVE =df_AAVE.merge(TVL_AAVE, left_on='time', right_on='date')
# df_AAVE_existing =df_AAVE_existing.merge(Network_AAVE_existing, left_on = 'time', right_on='time')
# df_AAVE_existing.head()

# merge new  datasets
# df_AAVE_new =df_AAVE_new.merge(Network_AAVE_new, left_on = 'time', right_on='time')
df_AAVE_merge = pd.merge(Network_AAVE_new, df_AAVE_new, left_index=True, right_index=True)
# drop column time_y
df_AAVE_merge = df_AAVE_merge.drop(["time_y"], axis = 1)
# rename the column
df_AAVE_merge.rename(columns = {'time_x':'time'}, inplace = True)
df_AAVE_merge.head()
```
